### Understanding the Data Structure (load_fits() - Before Visualization:

##### 1.) What is the shape of the image array, and what does each dimension represent?  
The shape (dimensions) of the image array is (12200, 8600). 12,200 represents the rows,  8,600  
represents the columns. This means the image consists of 12,200 rows and 8,600 columns of pixels.  
Each pixel is a measurement from the detector at that spatial location in the sky.  

##### Important Note:  
The FITS file (hdul.info()) reports dimensions as (8600, 12200) -> (columns, rows)  
NumPy arrays (image_data.shape) store the dimensions of the image as (12200, 8600) -> (rows, columns)  
This difference occurs because FITS follows an astronomical image convention, while NumPy follows a matrix convention.  

##### Extra Insight:  
In Numpy, array[row, column] so image_data[y, x]  
image_data[y, x] where y is the vertical position (row), and x is the horizontal position (column)  
This will matter later for image alignment and specific algorithms for detection and mapping.  
12200 x 8600 = 104,920,000 ~ 105 megapixels

##### 2.) Does each element correspond to a photon count, intensity value, or processed brightness value?  
Each element in the image array is a pixel value, which represents the measured signal recorded by the detector  
at that location on the sensor of the telescope during exposure. The pixel value, which is at a specific row and  
column in the array, corresponds to the intensity of detected photons in that region of the sky. The telescope also  
detects noise and background sky signal as well.  
##### Extra Insight:  
Because stars occupy such a small fraction of the image area, most pixels in a image of the night sky are measuring  
empty sky. This is important for stacking and background estimation later.


##### 3.) Why is the data stored as float32 instead of integers?  
This is for the purpose of mathematical equations and operations. Because image processing involves  
a ton of math, floats are better than integers because floating points can represent fractional values.  
Values that cannot be stored in an integer, can be stored perfectly in a float.  
##### Important:  
Scientific image processing involves complex arithmetic operations and scaling during calibration and  
enhancement. With integer values, subtraction could produce negative values, division would produce fractions,  
and rounding errors could pop up. Floating-points work to preserve numerical precision.

##### 4.) What would happen if calibration operations were performed on integer data instead?  
Astronomical calibration consists of operations that can produce fractional, and negative values,  
along with small precision differences. Because integer values cannot represent fractions, results  
would be rounded. This may result in key losses of precision and incorrect results if calibration  
is done with integer data. Across a large set of data, errors start to compound which is not ideal.

### Understanding Pixel Statistics:

##### 1.) Why might the minimum pixel value be negative?  
During the calibration process, when data is converted into floating-point numbers, negative values  
can be produced. These negative values represent noise and signal. Subtracting dark or bias frames  
from "light" frames can result in negative values.

##### 2.) Why might the maximum value be very large?  
Due to the fact that the sensor is capturing very faint objects and also very bright  
objects in the same frame. This huge difference will be observed after performing mathematics  
in the calibration process.

##### 3.) What does the distribution of pixel values likely look like?  
The distribution is most likely skewed, due to the fact that a large majority of the sensor  
pixels will be captured near the black-background of the night sky. A small amount of pixels  
will represent the bright objects in the night sky, such as the stars.

##### 4.) Why would most pixels correspond to background sky instead of stars?  
Because the night sky is primarily almost all empty space. Stars represent a tiny fraction  
of the sky, meaning most of the photons hitting the camera sensor are coming from the dark  
background of the night sky.

##### 5.) What does a single pixel physically represent in an astronomical detector?  
A pixel represents the number of photons detected from a tiny region of the night sky,  
during exposure time. A pixel can be modified by detector noise and instrument effects.

### Understanding Noise and Signal:

##### 1.) What sources of noise might exist in these raw images?  
The background of the night sky, cosmic rays (hot pixels), light pollution, read noise, etc.  
Cosmic rays give off high-energy particles which can strike the sensor creating bright pixels.  
Unwanted light from artificial sources, and electronic noise from the camera's circuitry when  
converting the analog charge in pixels into digital data.

##### 2.) What sources of signal are present?  
Photons from stars, nebulae, galaxies and any other astrophysical objects hitting the camera sensor,  
which is the desired data. Background noise such as environmental or instrumental interference, which  
is not the desired data.

##### 3.) Why are most pixels expected to be near background levels?  
Due to the fact that a vast majority of the photons hitting the camera sensor are coming  
from the massive empty background of space.
